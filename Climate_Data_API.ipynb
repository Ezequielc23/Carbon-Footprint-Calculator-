```python
def car_co2(make, motor, dist):
  url = "https://pre.backend.calculator.climatetrade.net/api/v1/calculate/calculate_cars_co2/?make_model={}&motor_type={}&km={}".format(make,motor,dist)
  car_co2 = requests.get(url)
  print('make and model:',make,' motor type:',motor,' distance in km:',dist)
  return car_co2.json()

def hotel_co2(stars, rooms, days, country):
  url = "https://pre.backend.calculator.climatetrade.net/api/v1/calculate/calculate_hotel_co2/?stars={}&rooms={}&days={}&country={}".format(stars,rooms,days,country)
  hotel_co2 = requests.get(url)
  print('stars:',stars,' rooms:',rooms,' stay length: (in days)',days,' country:',country)
  return hotel_co2.json()

def mcc_list(limit, offset):
  url = 'https://pre.backend.calculator.climatetrade.net/api/v1/retail-payments/?limit={}&offset={}'.format(limit,offset)
  mcc_list = requests.get(url)
  print('# of results to return/page:',limit,'\nThe initial index from which to return the results:',offset,'\n')
  return mcc_list.json()

def mcc_single(mcc):
  url = 'https://pre.backend.calculator.climatetrade.net/api/v1/retail-payments/{}/'.format(mcc)
  mcc = requests.get(url)
  print('Merchant Category Code: (mcc)',mcc,'\n')
  return mcc.json()

def calc_retail(amount,country,currency,mcc):
  url = "https://pre.backend.calculator.climatetrade.net/api/v1/retail-payments/calculate/"
  headers = CaseInsensitiveDict()
  headers["accept"] = "application/json"
  headers["Content-Type"] = "application/json"
  headers["X-CSRFToken"] = "INSERT TOKEN"

  data = {  "amount": 100,  "country": "india",  "currency": "USD",  "mcc": 3009}
  data["amount"] = amount
  data["country"] = country
  data["currency"] = currency
  data["mcc"] = mcc
  data = str(data)
  data = data.replace("'","\"")
  
  print(data)
  resp = requests.post(url, headers=headers, data=data)
  return resp.json()

def auth_token(username,password):
  url = 'https://pre.backend.climatetrade.net/api/token-auth/'
  header =  {"email": "", "password": ""}
  header["email"] = username
  header["password"] = password
  token_auth = requests.post(url,json=header)
  token = token_auth.json()
  token = dict(token)
  token = token['token']
  print('authenticated token: ')
  return token

def verify_token(token):
  url = 'https://pre.backend.climatetrade.net/api/token-verify/'
  header = {"token":""}
  header["token"] = token
  token_ver = requests.post(url,json=header)
  print('verified token: ')
  return token_ver.json()
  

def proj_info(proj_id):
  url = "https://pre.backend.climatetrade.net/api/v1/projects/"
  p = '{}'.format(proj_id)
  url = url+p+'/'

  headers = {"accept":"","authorization":"","X-CSRFToken":""}
  headers["accept"] = "application/json"
  headers["authorization"] = "INSERT AUTH HERE"
  headers["X-CSRFToken"] = "INSERT TOKEN HERE"
  resp = requests.get(url, headers=headers)

  return resp.json()

```
