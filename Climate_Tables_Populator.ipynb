```python
transact_carbon = spark.sql('select tc.Transaction_ID, tc.Carbon_Kg, tc.Confidence_Level, tc.Offset_Estimate_Amt from clm.transaction_carbon as tc join clm.transaction as t on t.Transaction_ID = tc.Transaction_ID where Status = "Cleared" and Carbon_Calc_f = "Y"')
transact_carbon = transact_carbon.toPandas()

data = {}
total = []

for row in api_response:
  data['Carbon_Kg'] = row['kg']
  data['Transaction_ID'] = row['transaction_id']
  total.append(data.copy())
  
total = pd.DataFrame(total)
total.sort_values("Transaction_ID", inplace = True)
transact_carbon.sort_values("Transaction_ID", inplace = True)
transact_c = transact_carbon[['Transaction_ID','Carbon_Kg']]
count=0

transact_c.sort_values("Transaction_ID", inplace = True)
numbers = total[["Carbon_Kg","Transaction_ID"]]
n = pd.DataFrame(numbers)
t = transact_c['Transaction_ID']
t = pd.DataFrame(t)
t.sort_values('Transaction_ID',inplace=True)

df3 = pd.merge(t, n)
df3['Confidence_Level'] = 'null'
df3['Offset_Estimate_Amt'] = 'null'

transaction_carbo=spark.createDataFrame(df3)
display(transaction_carbo)
```
